Uso de join y merge

1.¿Cuales son los profesores asignados a cada grupo?
2.¿Cuales son los estudiantes que tiene cada profesor?

```python
grupo['IDG'].unique()
```

```python
grupoProfesor=grupo.merge(profesor)
grupoProfesor.head(10)
```

```python
#nombre profrso id grupo,curso cantidad
grupoProfesor.sort_values('profesor',inplace=True)
ordenado=grupoProfesor[['profesor','IDG','curso','nestud']]
ordenado
```

¿Cuales son los estudiantes que tiene cada profesor, por grupo y programa académico?

```python
otroGrupo=listado.merge(grupoProfesor)
conEstudiantes=estudiante.merge(otroGrupo)
conEstudiantes.head(5)
conEstudiantes.sort_values(['profesor','IDG','programa','IDE'],inplace=True)
sabanaW=conEstudiantes[['profesor','IDG','programa','IDE']]
sabanaW
```

```python
filtro1=conEstudiantes.groupby(['profesor','programa']).agg(notaPromedio=('prom','mean'),numEstudiantes=('programa','count'))
filtro2=conEstudiantes.groupby(['profesor','programa']).agg({'prom':'mean','programa':'count'})
filtro1
```

```python
grupos=conEstudiantes.groupby(['profesor']).agg(grupo=('IDG','unique'))
grupos
```

```python
informe=conEstudiantes.pivot_table(index=['profesor','IDG'],columns='programa',values=['corte1','corte2'], aggfunc=np.mean)
informe
```
