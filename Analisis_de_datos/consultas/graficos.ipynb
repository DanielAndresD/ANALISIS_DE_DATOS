```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
rutaArchivo='examen2.xlsx' #Archivo a trabajar
ruta1='estudiante' #hoja del archivo
ruta2='profesor' #hoja del archivo
ruta3='grupo' #hoja del archivo
ruta4='listado' #hoja del archivo
ruta5='examen' #hoja del archivo
ruta6='cita' #hoja del archivo

estudiante= pd.read_excel(rutaArchivo,sheet_name=ruta1)
profesor=pd.read_excel(rutaArchivo,sheet_name=ruta2)
grupo=pd.read_excel(rutaArchivo,sheet_name=ruta3)
listado=pd.read_excel(rutaArchivo,sheet_name=ruta4)
examen=pd.read_excel(rutaArchivo,sheet_name=ruta5)
cita=pd.read_excel(rutaArchivo,sheet_name=ruta6)
```

```python
baseE=estudiante.merge(listado)
baseG=grupo.merge(profesor)
baseS=baseE.merge(baseG)
base1=baseS.merge(examen)
base=base1.merge(cita)
```
