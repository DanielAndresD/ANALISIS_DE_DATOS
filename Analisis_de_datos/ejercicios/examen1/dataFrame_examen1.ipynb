Recursos necesarios de trabajo

```python
rutaArchivo='examen1.xlsx' #Archivo a trabajar
hoja='examen1' #hoja del archivo
```

Vincular el recurso donde se encuentran almacenados los datos en un documento excel, así como ver la información de los campos registrados


```python
data= pd.read_excel(rutaArchivo,sheet_name=hoja)
#Eliminar campo de conteo
data=data.drop(data.columns[[0,1]],axis=1)
data.info()
```

```python
#Filtrar jornada día orden ascendente por docente
base2=data[data['jornada']=='Día'].sort_values(['profesor'],ascending=(True))
base2
```

```python
#Guardar por diferente tipo
base2.to_excel('jornadaDia.xlsx',sheet_name="Dia")
base2.to_csv('jornadaDia.csv',sep=",")
```
